```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Function to process and count leaves in a single image
def process_image(image_path, min_leaf_area=500):
  
    image = cv2.imread(image_path)

    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply GaussianBlur to reduce noise
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    # Apply a binary threshold to segment the plant from the background
    _, thresholded = cv2.threshold(blurred, 120, 255, cv2.THRESH_BINARY)

    # Find contours of the segmented plant region
    contours, _ = cv2.findContours(thresholded, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Filter contours to remove noise (optional, based on area)
    valid_contours = [cnt for cnt in contours if cv2.contourArea(cnt) > min_leaf_area]

    # Return the number of valid contours (leaves)
    return len(valid_contours)

# List of image filenames (for example, images taken over multiple days)
image_folder = 'INPUT'
image_filenames = sorted(os.listdir(image_folder))  # Assuming filenames are sorted in order (e.g., day1.jpg, day2.jpg, ...)

# Prepare lists to store leaf counts and corresponding days
leaf_counts = []
days = []

# Process each image and count leaves
for image_filename in image_filenames:
    # Generate the full path to the image
    image_path = os.path.join(image_folder, image_filename)
    
    # Extract the day from the image filename (assuming format 'dayX.jpg')
    day = image_filename.split('.')[0]  # This splits 'dayX.jpg' into 'dayX' and takes 'X'
    
    # Process the image and get the leaf count
    leaf_count = process_image(image_path)
    
    # Append the results to the lists
    leaf_counts.append(leaf_count)
    days.append(day)

# Plot the leaf counts over time (days)
plt.figure(figsize=(10, 6))
plt.plot(days, leaf_counts, marker='o', linestyle='-', color='b', label='Leaf Count')
plt.xlabel('Days')
plt.ylabel('Leaf Count')
plt.title('Leaf Count over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()


for day, count in zip(days, leaf_counts):
    print(f'{day}: {count} leaves')
```
